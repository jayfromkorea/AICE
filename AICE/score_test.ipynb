```python
names = [make_korean_name() for _ in range(100)]
print(names)
```

```python
di = {
    '번호' : np.arange(1, 101), # 1 - 100까지 순차적인 값들의 배열 생성
    '이름' : names,
    '학년' : 3,
    '반' : np.random.randint(1, 6, size=100),   # 1 - 5까지 100개의 데이터 생성
    '국어' : np.clip(np.round(np.random.normal(40, 10, size=100), 2), 0, 100),    # 정규분포임
    '수학' : np.clip(np.round(np.random.normal(60, 10, size=100), 2), 0, 100),    # np.clip은 범위 밖으로 나가는 데이터를 잘라줌
    '영어' : np.clip(np.round(np.random.normal(70, 10, size=100), 2), 0, 100),
    '역사' : np.clip(np.round(np.random.normal(50, 10, size=100), 2), 0, 100),
}

df = pd.DataFrame(di)
df
```

```python
# 필터링하기
cond = df['역사'] > 100.0
df[cond]
```

```python
# 국어 60점 이상, 영어 80점 이상인 학생들만 가져오기
cond = (df['국어'] >= 60.0) & (df['영어'] >= 80.0)
df[cond]
```

```python
# 데이터 추가하기
# 총점이라는 데이터를 생성하여 컬럼을 추가함
df['총점'] = df['국어'] + df['수학'] + df['영어'] + df['역사']
df['평균'] = (df['총점'] / 4).round(2)
df.head()
```

```python
# 100명 전체의 등수 구하기
df['석차'] = df['총점'].rank(ascending=False).astype(int)
df
```

```python
# 데이터 정렬하기
df.sort_values(by='석차')
```
