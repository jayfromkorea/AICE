# Series 

- 표 데이터의 컬럼(열)을 연상시키는 오브젝트


```python
li = [10, 20, 30, 40, 50]
s1 = pd.Series(li)
s1
```

```python
label = ['a', 'b', 'c', 'd', 'e']
s1 = pd.Series(li, index=label)
s1
```

```python
di = {
    'a':10,
    'b':20,
    'c':30,
    'd':40,
    'e':50
}

s1 = pd.Series(di)
s1
```

```python
print(s1['a'])  # 얘랑
print(s1.loc['b'])  # 얘항은 같음
# print(s1[1]) # 이건 안됨
print(s1.iloc[1])   # 절대 행 번호를 지정하여 데이터에 접근함
print(s1.iloc[-1])
```

```python
print(s1['a': 'c'])
print(s1['b':])
print(s1.iloc[1:4])

# Fancy Indexing
print(s1[['a', 'c', 'e']])
print(s1.iloc[[0, 2, 4]])
```

```python
# Series 객체에서  자주 사용하는 속성
print(s1.values)    # 값들을 ndarray로 보여줌
print(s1.index)     # 인덱스 객체를 반환
print(s1.array)     # Numpy Extension Array로 반환
```

```python
# 값들의 수정
s1['a'] = 100
print(s1)

s1.iloc[0] = 200
s1
```

```python
# 추가
s1['f'] = 90    # 없는 키를 사용하여 값 추가
s1
```

```python
# 삭제
# s1에서는 삭제처리가 되지 않고, 삭제된 copy를 반환함
s2 = s1.drop('c')
print(s1)
print(s2)
```

```python
# inplace=True를 지정하면 객체 내부의 데이터를 삭제처리한다.
s1.drop('c', inplace=True)
s1
```

```python
# 한번에 여러개의 데이터를 삭제해줌
s2 = s1.drop(['a', 'b'])
s2
```

```python
# 사칙연산을 브로드캐스팅 방식으로 지원한
# 이때 내부의 데이터를 변경하지 않고 연산된 새로운 Series객체를 반환한다.
print(s1 * 10)
print(s1 - 5)
```

```python
# Filtering
cond = s1 <= 40
print(cond, type(cond))
s1[cond]
```

# DataFrame
- 2차원 형태의 데이터 set을 다루기 위한 오브젝트

```python
di = {
    'name':['Kim', 'Lee', 'Part', 'Choi'],
    'age':[30, 40, 50, 60],
    'score':[23, 43, 56, 34]
}

df = pd.DataFrame(di)
df
```

```python
di = [
    {'name':'kim', 'age':30, 'score':40},
    {'name':'lee', 'age':40, 'score':50},
    {'name':'park', 'age':50, 'score':60},
    {'name':'choi', 'age':60, 'score':70}
]

df = pd.DataFrame(di)
df
```
